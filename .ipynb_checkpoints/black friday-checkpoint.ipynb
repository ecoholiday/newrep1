```python
fmri = sns.load_dataset("fmri") 
fmri
```

```python
sns.lineplot(x="timepoint", y="signal",
             hue="region",style='event',
             data=fmri)
```

```python
sns.relplot(x="timepoint", y="signal", kind="line",hue='region',style='event', data=fmri)
```

```python
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri,estimator=None)
```

```python
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri) # by default it does with 95% cofidence interval
```

```python
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri,ci=None)
```

```python
sns.relplot(x="timepoint", y="signal", kind="line",hue='region',style='event',
            markers=True,dashes = False,data=fmri)
```

```python
sns.relplot(x="timepoint", y="signal", hue="region",
            units="subject", estimator=None,
            kind="line",data=fmri[fmri.event =='stim'])  # caould use units to avoid it showing in legends
```

### facets are possible by using col='' and row = ''

```python
sns.relplot(x="timepoint", y="signal",hue = 'subject',
            col="region",row = 'event'
            ,kind="line",data=fmri)
```
